```python
#tfs = os.listdir("../RESULTS_INDIV_TFS_REFINED_SUBSAMPLE_5000/")
#tfs = os.listdir("../RESULTS_INDIV_TFS_REFINED_SUBSAMPLE_1000/")
#tfs = os.listdir("../RESULTS_INDIV_TFS_REFINED_SUBSAMPLE_500/")
#tfs = os.listdir("../RESULTS_INDIV_TFS_REFINED_SUBSAMPLE_250/")
#tfs = os.listdir("../RESULTS_INDIV_TFS_REFINED_SUBSAMPLE_125/")
tfs = os.listdir("../RESULTS_INDIV_TFS_REFINED_SUBSAMPLE_50/")

tf_data = {}
for tf in tfs:
    
    #if tf != "CTCF": 
    #    continue
    
    tl_data = {}
    no_tl_data = {}
    
    for i in range(1,6):
        
        tl_data[i] = []
        no_tl_data[i] = []
        
        for j in range(1,6):
            pkl_file = open("../RESULTS_INDIV_TFS_REFINED_SUBSAMPLE_50/"+tf+"/iterat_TL_"+
                            str(i)+"_" + str(j) +"/mccoef.pkl", 'rb')
            mccoef_TL = pickle.load(pkl_file)
            pkl_file.close()
            
            pkl_file = open("../RESULTS_INDIV_TFS_REFINED_SUBSAMPLE_50/"+tf+"/iterat_noTL_"+
                            str(i)+"_" + str(j) +"/mccoef.pkl", 'rb')
            mccoef_noTL = pickle.load(pkl_file)
            pkl_file.close()
            
            tl_data[i].append(list(mccoef_TL.values())[0])
            no_tl_data[i].append(list(mccoef_noTL.values())[0])
            
    single_data = {}
    single_data["TL"] = tl_data
    single_data["No_TL"] = no_tl_data
    
    tf_data[tf] = single_data
    
    
new_model_TL = {}
new_model_no_TL = {}

for i in range(1,11):
    pkl_file = open("../RESULTS_50_SORTED/iterat_TL_"+str(i)+"/mccoef.pkl", 'rb')
    mccoef_new_model_TL = pickle.load(pkl_file)
    pkl_file.close()

    pkl_file = open("../RESULTS_50_SORTED/iterat_noTL_"+str(i)+"/mccoef.pkl", 'rb')
    mccoef_new_model_no_TL = pickle.load(pkl_file)
    pkl_file.close()
    
    for TF in tfs:
        if TF not in new_model_TL.keys() and TF in mccoef_new_model_TL.keys():
            new_model_TL[TF] = []
            new_model_TL[TF].append(mccoef_new_model_TL[TF])
        elif TF in mccoef_new_model_TL.keys():
            new_model_TL[TF].append(mccoef_new_model_TL[TF])
        ##################################################################
            
        ##################################################################
        if TF not in new_model_no_TL.keys() and TF in mccoef_new_model_no_TL.keys():
            new_model_no_TL[TF] = []
            new_model_no_TL[TF].append(mccoef_new_model_no_TL[TF])
        elif TF in mccoef_new_model_no_TL.keys():
            new_model_no_TL[TF].append(mccoef_new_model_no_TL[TF])
            
new_model_TL = pd.Series(new_model_TL) 
new_model_no_TL = pd.Series(new_model_no_TL) 
```
