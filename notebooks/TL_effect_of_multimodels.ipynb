```python
from tools import *
from models import *
import plotly.graph_objects as go
import plotly.figure_factory as ff
from Bio.SeqUtils import GC
from Bio import SeqIO
import os
from random import sample
from plotly.subplots import make_subplots
import pickle
from scipy import stats
from collections import Counter

plt.ioff()

import warnings
warnings.filterwarnings('ignore')
```

```python
#RECORDING THE PERFORMANCE
TFs = ["JUND", "HNF4A", "MAX", "SP1", "SPI1"]

results_bm_multi = {}
results_bm_small_multi = {}

results_cof_multi = {}
results_cof_small_multi = {}

results_str_multi = {}
results_str_small_multi = {}

results_bestcor_multi = {}
results_bestcor_small_multi = {}

results_beststr_multi = {}
results_beststr_small_multi = {}

for TF in TFs:
    results_bm_multi[TF] = []
    results_bm_small_multi[TF] = []
    
    results_cof_multi[TF] = []
    results_cof_small_multi[TF] = []
    
    results_bestcor_multi[TF] = []
    results_bestcor_small_multi[TF] = []
    
    results_str_multi[TF] = []
    results_str_small_multi[TF] = []
    
    results_beststr_multi[TF] = []
    results_beststr_small_multi[TF] = []
    
    for i in range(1,6):
    
        pkl_file = open("../RESULTS_BM_MULTIMODEL/"+
                        TF+"_50_TFs_"+str(i)+"/mccoef.pkl", 'rb')
        mccoef_true = pickle.load(pkl_file)
        pkl_file.close()
        results_bm_multi[TF].append(list(mccoef_true.values())[0])
        
        pkl_file = open("../RESULTS_COFACTOR_MULTIMODEL/"+
                        TF+"_50_TFs_"+str(i)+"/mccoef.pkl", 'rb')
        mccoef_true = pickle.load(pkl_file)
        pkl_file.close()
        results_cof_multi[TF].append(list(mccoef_true.values())[0])
        
        pkl_file = open("../RESULTS_STRING_MULTIMODEL/"+
                        TF+"_50_TFs_"+str(i)+"/mccoef.pkl", 'rb')
        mccoef_true = pickle.load(pkl_file)
        pkl_file.close()
        results_str_multi[TF].append(list(mccoef_true.values())[0])
        
        pkl_file = open("../RESULTS_BESTCOR_MULTIMODEL/"+
                        TF+"_50_TFs_"+str(i)+"/mccoef.pkl", 'rb')
        mccoef_true = pickle.load(pkl_file)
        pkl_file.close()
        results_bestcor_multi[TF].append(list(mccoef_true.values())[0])
        
        pkl_file = open("../RESULTS_BESTSTR_MULTIMODEL/"+
                        TF+"_50_TFs_"+str(i)+"/mccoef.pkl", 'rb')
        mccoef_true = pickle.load(pkl_file)
        pkl_file.close()
        results_beststr_multi[TF].append(list(mccoef_true.values())[0])
        
        #################################################################
        
        pkl_file = open("../RESULTS_BM_MULTIMODEL/"+
                        TF+"_5_TFs_"+str(i)+"/mccoef.pkl", 'rb')
        mccoef_true = pickle.load(pkl_file)
        pkl_file.close()
        results_bm_small_multi[TF].append(list(mccoef_true.values())[0])
        
        pkl_file = open("../RESULTS_COFACTOR_MULTIMODEL/"+
                        TF+"_5_TFs_"+str(i)+"/mccoef.pkl", 'rb')
        mccoef_true = pickle.load(pkl_file)
        pkl_file.close()
        results_cof_small_multi[TF].append(list(mccoef_true.values())[0])
        
        pkl_file = open("../RESULTS_STRING_MULTIMODEL/"+
                        TF+"_5_TFs_"+str(i)+"/mccoef.pkl", 'rb')
        mccoef_true = pickle.load(pkl_file)
        pkl_file.close()
        results_str_small_multi[TF].append(list(mccoef_true.values())[0])
        
        pkl_file = open("../RESULTS_BESTCOR_MULTIMODEL/"+
                        TF+"_5_TFs_"+str(i)+"/mccoef.pkl", 'rb')
        mccoef_true = pickle.load(pkl_file)
        pkl_file.close()
        results_bestcor_small_multi[TF].append(list(mccoef_true.values())[0])
        
        pkl_file = open("../RESULTS_BESTSTR_MULTIMODEL/"+
                        TF+"_5_TFs_"+str(i)+"/mccoef.pkl", 'rb')
        mccoef_true = pickle.load(pkl_file)
        pkl_file.close()
        results_beststr_small_multi[TF].append(list(mccoef_true.values())[0])
        
                
results_bm_multi = pd.Series(results_bm_multi)
results_bm_small_multi = pd.Series(results_bm_small_multi)

results_cof_multi = pd.Series(results_cof_multi)
results_cof_small_multi = pd.Series(results_cof_small_multi)

results_str_multi = pd.Series(results_str_multi)
results_str_small_multi = pd.Series(results_str_small_multi)

results_bestcor_multi = pd.Series(results_bestcor_multi)
results_bestcor_small_multi = pd.Series(results_bestcor_small_multi)

results_beststr_multi = pd.Series(results_beststr_multi)
results_beststr_small_multi = pd.Series(results_beststr_small_multi)
```

```python
x_models = [[tf]*5 for tf in results_bm_multi.index]
x_models = [item for sublist in x_models for item in sublist]
```

```python
multimodel_bm_tfs = []
for tf in results_bm_multi.index:
    multimodel_bm_tfs = multimodel_bm_tfs + results_bm_multi[tf]
    
multimodel_cof_tfs = []
for tf in results_bm_multi.index:
    multimodel_cof_tfs = multimodel_cof_tfs + results_cof_multi[tf]
    
multimodel_str_tfs = []
for tf in results_bm_multi.index:
    multimodel_str_tfs = multimodel_str_tfs + results_str_multi[tf]
    
multimodel_bestcor_tfs = []
for tf in results_bm_multi.index:
    multimodel_bestcor_tfs = multimodel_bestcor_tfs + results_bestcor_multi[tf]
    
##############################################################################

multimodel_bm_small_tfs = []
for tf in results_bm_multi.index:
    multimodel_bm_small_tfs = multimodel_bm_small_tfs + results_bm_small_multi[tf]
    
multimodel_cof_small_tfs = []
for tf in results_bm_multi.index:
    multimodel_cof_small_tfs = multimodel_cof_small_tfs + results_cof_small_multi[tf]
    
multimodel_str_small_tfs = []
for tf in results_bm_multi.index:
    multimodel_str_small_tfs = multimodel_str_small_tfs + results_str_small_multi[tf]
    
multimodel_bestcor_small_tfs = []
for tf in results_bm_multi.index:
    multimodel_bestcor_small_tfs = multimodel_bestcor_small_tfs + results_bestcor_small_multi[tf]
```

```python
fig = go.Figure()

fig.add_trace(go.Box(
    y=multimodel_tfs,
    x=x_models,
    name='50 TFs, multimodel',
    marker_color='red'
))
fig.add_trace(go.Box(
    y=multimodel_small_tfs,
    x=x_models,
    name='5 TFs, multimodel',
    marker_color='#FF4136'
))

fig.update_layout(
    yaxis_title='MCC',
    boxmode='group' # group together boxes of the different traces for each value of x
)

fig.update_layout(title='Comparison of TL with different multimodels (BESTCOR)',
                 plot_bgcolor='rgba(0,0,0,0)', paper_bgcolor='rgba(0,0,0,0)')

fig.update_xaxes(showline=True, linewidth=2, linecolor='black')
fig.update_yaxes(showline=True, linewidth=2, linecolor='black')

fig.show()
```
